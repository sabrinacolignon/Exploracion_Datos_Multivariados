### 3. ARREGLO DE DATOS

##### DECLARACION DE VARIABLES
La fila de abreviaturas de tres (o cuatro) letras describen el tipo de cada variable:

int significa números enteros.

dbl significa dobles o números reales.

chr significa vectores de caracteres o cadenas.

dttm significa fecha-hora (una fecha + una hora).

lgl significa vectores lógicos que contienen solo TRUEo FALSE.

fctr significa factores, que R usa para representar variables categóricas con valores posibles fijos.

date significa fechas.

Vemos que la especificación del tipo de dato no se encuentra cargado correctamente, debemos tener en cuenta que en R si un dato es discreto debe estar cagado como integer, si es continuo debe estar cargado como numeric, si es cualitativo categorico debe ser un factor, si es dicotómico debe ser logico, podemos cambiar esto utilizando

Función	Descripción
as.numeric()	Cambia la naturaleza de un vector hacia clase numérica (decimales).
as.integer()	Cambia la naturaleza de un vector hacia clase numérica (enteros).
as.character()	Cambia la naturaleza de un vector hacia clase texto o carácter.
as.logical()	Cambia la naturaleza de un vector hacia clase lógica (TRUE o FALSE).
as.factor()     Cambia la naturaleza de un vector hacia clase factor (categorias)

Escribir>
Nombre$Variable<-as.numeric(Nombre$Variable)

Ver más coerciones https://irwingss.github.io/intro_r_ciencias/coerci%C3%B3n-de-variables.html

### MANIPULACION DE DATOS CON PAQUETE DPLYR

Habiendo revisado el contenido de un dataframe, comenzamos a hacernos idea de los ajustes que necesita para que los datos tomen el formato que necesitamos. Estos ajustes pueden ser correcciones (por ejemplo, de errores de tipeo cuando se cargaron los datos), la creación de nuevas variables derivadas de las existentes, o un reordenamiento de los datos para simplificar nuestro trabajo.

rename(): renombrar variables por ejemplo

Nombre <-rename(Nombre, NombreVariableNuevo = varible1,NombreVariableNuevo = varible2 )

select(): seleccionar -elegir- columnas por su nombre 
Hay una serie de funciones auxiliares que puede utilizar dentro de select():

starts_with("abc"): coincide con los nombres que comienzan con "abc".

ends_with("xyz"): coincide con los nombres que terminan con "xyz".

contains("ijk"): coincide con los nombres que contienen "ijk".

matches("(.)\1"): selecciona variables que coinciden con una expresión regular. Este coincide con cualquier variable que contenga caracteres repetidos. Aprenderá más sobre las expresiones regulares en cadenas .

por ejemplo
Nombre <-select(Nombre, varible1, varible2)

filter(): filtrar, es decir quedarse sólo con las filas que cumplan cierta condición:
Los operadores de comparación en R son los siguientes:
< menor que

> mayor que

>= mayor o igual que

== igualdad

!= no igual

por ejemplo
Nombre1 <-filter(Nombre, varible1=="caracteristica") #El dataset Nombre1 contendra una subconjunto de datos donde la variable1 toma el valor caracteristica

mutate(): mutar -cambiar- un dataframe, modificando el contenido de sus columnas o creando columnas (es decir, variables) nuevas

case_when() usa expresiones lógicas para asignar valores a las variables

group_by() agrupa un conjunto de filas seleccionado en un conjunto de filas de resumen de acuerdo con los valores de una o más columnas o expresiones.

summarise(): producir sumarios -un valor extraído de muchos, por ejemplo el promedio- con el contenido de las columnas

arrange(): ordenar las filas de acuerdo a su contenido o algún otro índice

Combinar datos con R Dplyr
