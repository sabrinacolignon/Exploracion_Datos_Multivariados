```r
Frec_Prom_Matematica <- fdt(filter(Diamante_limpio, asignatura=="MATEMÁTICA")$nota, breaks="Sturges",na.rm=TRUE)
# calcula la distribución de frecuencias utilizando la regla Sturge
```

```r
Diamante_limpio_matematica <- filter(Diamante_limpio, asignatura=="MATEMÁTICA")
```

```r
#intervalos y frecuencias con dplyr
Frec_Prom_Matematica <- Diamante_limpio_matematica[!is.na(Diamante_limpio_matematica$nota),] %>% 
                          mutate(notas = cut(nota, breaks = 6), na.rm=TRUE) %>% 
                          group_by(notas) %>% 
                          summarise(frequency = n()) %>% 
mutate(acumulado = cumsum(frequency))
```

```r
Frec_Prom_Matematica #nos brinda una tabla con los calculos de la distribución de frecuencias. breaks=c('Sturges', 'Scott', 'FD')
```

start: left endpoint of the first class interval.

end: right endpoint of the last class interval.

h: class interval width.

right: right endpoints open (default = FALSE).

```r
dst_1 <- filter(Diamante_limpio, Curso == 1)
dst_2 <- filter(Diamante_limpio, Curso == 2)
dst_3 <- filter(Diamante_limpio, Curso == 3)
dst_4 <- filter(Diamante_limpio, Curso == 4)
dst_5 <- filter(Diamante_limpio, Curso == 5)
dst_6 <- filter(Diamante_limpio, Curso == 6)
t <-length(unique(Diamante_limpio$Documento)) 
proporciones <- c(length(unique(dst_1$Documento))/t,length(unique(dst_2$Documento))/t,length(unique(dst_3$Documento))/t,
                length(unique(dst_4$Documento))/t,length(unique(dst_5$Documento))/t,length(unique(dst_6$Documento))/t)

cur <- c(1,2,3,4,5,6)
propcur <- data.frame("Curso"=cur,
                     "Proporcion"=proporciones)
propcur
```

```r
privada <- filter(Diamante_limpio, EsPrivada == "S")
publica <- filter(Diamante_limpio, EsPrivada == "N")
total <- length(unique(Diamante_limpio$Documento))
estudiantes_priv  <- length(unique(privada$Documento))
estudiantes_pub <- length(unique(publica$Documento))
proporciones <- c(estudiantes_priv/total,estudiantes_pub/total)
tabla <- data.frame("Gestion" = c("Privada","Publica"),
                   "Proporcion" = proporciones)
tabla
```
