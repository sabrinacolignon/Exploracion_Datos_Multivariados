```r
library(readxl)
library(dplyr)
library(scales)
```

```r
options(warn=-1)
Diamante <- read_excel("C:/Users/Sabrina/Downloads/Exploracion_d_multivariados/bd_primaria.xlsx", sheet = 1)
```

```r
prim_len_1 <- filter(Diamante, Curso== 1 & Materia=="LENGUA")
```

```r
prim_len_1[!is.na(prim_len_1$Promedio),] %>% 
  mutate(Promedio = cut(Promedio, breaks = 5), na.rm=TRUE) %>% 
  group_by(Promedio) %>% 
  summarise(frequency = n()) %>% 
  mutate(acumulado = cumsum(frequency))

boxplot(prim_len_1[!is.na(prim_len_1$Promedio),]$Promedio,
        main ="Boxplot para la asignatura Lengua de 1°grado de Primaria",
        horizontal = FALSE,
        sub = "Fuente de elaboración propia",
        whisklty = 2,       # Tipo línea bigotes
        whisklwd = 2,       # Ancho línea bigotes
        whiskcol = "red",   # Color bigotes
        staplelty = 3,      # Tipo línea barras
        staplelwd = 2,      # Ancho línea barras
        staplecol = "blue",  #Color barras
        col ="orange") #color caja

# Añadir línea con la media
abline(h = mean(prim_len_1[!is.na(prim_len_1$Promedio),]$Promedio), col = 4, lwd = 2) # Línea completa

# Añadir punto con la media
points(mean(prim_len_1[!is.na(prim_len_1$Promedio),]$Promedio), col = 3, pch = 19)
```

```r
hist(x = prim_len_1$Promedio,
     probability = TRUE, 
     col="orange",
     breaks = c(6,6.8,7.6,8.4,9.2,10), #cantidad de barras que quiero
     xaxp = c(6, 10, 5), #acomoda el eje x para ver bien los valores
     ylim = c(0, 0.5),
     xlab="notas",
     ylab="frecuencia relativa",
     sub = "Fuente de elaboración propia",
     main ="Histograma para la asignatura Lengua de 1°grado de Primaria")
```
